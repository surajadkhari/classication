```python
import requests
import pandas as pd
from requests.auth import HTTPBasicAuth
```

```python
api_url = "http://localhost:1251/api/Person"
username = "support@uniphi.com.au"
password = "Luna@9860"
response = requests.get(api_url, auth=HTTPBasicAuth(username, password), timeout=10)
print("Status Code:", response.status_code) 
print("Raw JSON Preview:")
print(response.json())  
```

```python
data=response.json()
df = pd.DataFrame(data)
df.info
```

```python
def extract_discipline(custom_fields):
    if isinstance(custom_fields, list):
        for field in custom_fields:
            if field.get("Name") == "Discipline":
                return field.get("Value")
    return None
# Apply custom function to extract Discipline
df["Discipline"] = df["CustomFields"].apply(extract_discipline)
old_classification_data=df[["ID","ClassificationID","ClassificationCode","Classification","Discipline","Position"]].drop_duplicates()
old_classification_data.head()
old_classification_data.to_csv("uniphi_person_classification.csv", index=False)
```
